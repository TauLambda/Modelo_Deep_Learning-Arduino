```python
import cv2
import os

# Ruta al modelo de clasificador en cascada
cascade_model_path = 'cascade.xml'

# Directorio que contiene las imágenes de vehículos
image_directory = 'C:/Users/Admin/Downloads/Dataset placas completo/Dataset placas'

# Carpeta para las imágenes detectadas como placas
output_directory = 'C:/Users/Admin/Downloads/Placas Detectadas1'

if not os.path.exists(output_directory):
    os.makedirs(output_directory)
# Cargar el modelo de clasificador en cascada
plate_cascade = cv2.CascadeClassifier(cascade_model_path)

# Contadores
total_images = 0
images_with_plates_count = 0

# Procesar cada imagen en el directorio
for filename in os.listdir(image_directory):
    if filename.endswith('.jpg') or filename.endswith('.png') or filename.endswith('.jpeg'):
        total_images += 1
        image_path = os.path.join(image_directory, filename)
        image = cv2.imread(image_path)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Realizar la detección de placas en la imagen
        plates = plate_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        # Si se detecta una placa en la imagen
        if len(plates) > 0:
            images_with_plates_count += 1

            # Marcar el área de la placa con un rectángulo verde
            for (x, y, w, h) in plates:
                cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

            # Guardar la imagen detectada como placa en la carpeta de salida
            output_filename = f"placa_{filename}"
            output_path = os.path.join(output_directory, output_filename)
            cv2.imwrite(output_path, image)

# Calcular el porcentaje de imágenes con placas detectadas
if total_images > 0:
    detection_percentage_m1 = (images_with_plates_count / total_images) * 100
else:
    detection_percentage_m1 = 0

# Imprimir los resultados
print(f"Total de imágenes procesadas: {total_images}")
print(f"Imágenes con placas detectadas: {images_with_plates_count}")
print(f"Porcentaje de imágenes con placas detectadas: {detection_percentage_m1:.2f}%")
```

```python
import cv2
import os
import numpy as np

def extraccion(img):
    # Load the image
    image = cv2.imread(str(img))

    # Check if the image is loaded successfully
    if image is None:
        print("Error: Image not loaded.")
        return None 
    else:
        # Resize the image to fit within the display window
        height, width = image.shape[:2]
        max_height = 800  # Set a maximum display height
        scale = max_height / height
        image_resized = cv2.resize(image, (int(width * scale), int(height * scale)))

        # Convert the resized image to grayscale
        gray = cv2.cvtColor(image_resized, cv2.COLOR_BGR2GRAY)

        # Apply Gaussian blur to reduce noise
        blurred = cv2.GaussianBlur(gray, (5, 5), 0)

        # Perform edge detection using Canny
        edges = cv2.Canny(blurred, 50, 150)

        # Find contours in the edged image
        contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Initialize a list to store potential license plate regions
        plate_regions = []

        # Iterate over detected contours
        for contour in contours:
            # Filter contours based on area
            if cv2.contourArea(contour) > 1000:
                epsilon = 0.03 * cv2.arcLength(contour, True)
                approx = cv2.approxPolyDP(contour, epsilon, True)

                # Check if the contour has 4 vertices (likely a rectangle)
                if len(approx) == 4:
                    plate_regions.append(approx)

        # Extract and show the detected plate(s)
        for i, region in enumerate(plate_regions):
            x, y, w, h = cv2.boundingRect(region)
            plate = image_resized[y:y+h, x:x+w]
            
            # Check if plate is None or its shape is (0, 0)
            if plate is not None and plate.shape != (0, 0):
                return plate

        return None  # Return None if no valid plate is found

# Ruta de la carpeta que contiene las imágenes
carpeta_imagenes  = 'C:/Users/Admin/Downloads/Dataset placas completo/Dataset placas'


# Ruta donde se guardarán las placas extraídas
ruta_salida = 'C:/Users/Admin/Downloads/Placas Detectadas4'

# Crear el directorio de salida si no existe
os.makedirs(ruta_salida, exist_ok=True)

# Contadores
placas_detectadas = 0
total_imagenes = 0

# Procesar todas las imágenes en la carpeta
for numero_imagen in range(1, 442):  # Suponiendo que tienes imágenes desde 0001.jpeg hasta 0303.jpeg
    nombre_imagen = f'{numero_imagen:04d}.jpeg'  # Formato de nombre de imagen, por ejemplo, 0001.jpeg
    ruta_imagen = os.path.join(carpeta_imagenes, nombre_imagen)

    # Cargar la imagen
    # imagen = cv2.imread(ruta_imagen)
    # Llamar a la función de extracción de placas
    placa_recortada = extraccion(ruta_imagen)

    # Actualizar el contador de imágenes procesadas
    total_imagenes += 1

    # Si no se encontró una placa en la imagen, placa_recortada será None
    if placa_recortada is not None:
        # Guardar la placa con el número de la imagen en el nombre
        nombre_salida = f'placa_{numero_imagen}.jpeg'
        ruta_salida_placa = os.path.join(ruta_salida, nombre_salida)
        cv2.imwrite(ruta_salida_placa, placa_recortada)

        placas_detectadas += 1
        print(f'Placa extraída de {nombre_imagen}')
    else:
        print(f'No se encontró una placa en {nombre_imagen}')

# Calcular el porcentaje de placas detectadas
porcentaje_placas_detectadas_m4 = (104 / total_imagenes) * 100

print(f'Proceso completado. {placas_detectadas} placas detectadas de {total_imagenes} imágenes ({porcentaje_placas_detectadas_m4:.2f}%).')
```

```python
import cv2
import os

# Ruta al modelo de clasificador en cascada
cascade_model_path = 'cascade.xml'

# Carpeta para las imágenes detectadas como placas
output_directory = 'C:/Users/Admin/Downloads/Placas Detectadas1'

if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Cargar el modelo de clasificador en cascada
plate_cascade = cv2.CascadeClassifier(cascade_model_path)

# Iniciar la captura de video desde la cámara (0 es la cámara predeterminada)
cap = cv2.VideoCapture(0)

# Contadores
total_frames = 0
frames_with_plates_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    total_frames += 1
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Realizar la detección de placas en el cuadro actual del video
    plates = plate_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Si se detecta una placa en el cuadro actual
    if len(plates) > 0:
        frames_with_plates_count += 1

        # Marcar el área de la placa con un rectángulo verde
        for (x, y, w, h) in plates:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Guardar el cuadro actual con la placa detectada en la carpeta de salida
        output_filename = f"frame_{total_frames}.jpg"
        output_path = os.path.join(output_directory, output_filename)
        cv2.imwrite(output_path, frame)

    # Mostrar el cuadro actual con la detección en tiempo real
    cv2.imshow('Plates Detected', frame)

    # Salir del bucle si se presiona la tecla 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar la captura y cerrar las ventanas al finalizar
cap.release()
cv2.destroyAllWindows()

# Calcular el porcentaje de cuadros con placas detectadas
if total_frames > 0:
    detection_percentage = (frames_with_plates_count / total_frames) * 100
else:
    detection_percentage = 0

# Imprimir los resultados
print(f"Total de cuadros procesados: {total_frames}")
print(f"Cuadros con placas detectadas: {frames_with_plates_count}")
print(f"Porcentaje de cuadros con placas detectadas: {detection_percentage:.2f}%")
```
